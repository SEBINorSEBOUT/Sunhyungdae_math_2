```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[5, 5], [5, 5]])
b = np.array([10, 20])

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# 각 선형 방정식을 그립니다.
for i in range(len(b)):
    # 각 방정식의 계수와 상수항을 가져옵니다.
    c1, c2 = A[i]
    c3 = b[i]

    # 두 번째 계수가 0이면 수직선을 그립니다.
    if c2 == 0:
        y = np.linspace(-20, 20, 100)
        x = c3 / c1 * np.ones_like(y)
        plt.plot(x, y, color='black')
    # 그 외에는 일반적인 직선을 그립니다.
    else:
        x = np.linspace(-20, 20, 100)
        y = (c3 - c1 * x) / c2
        plt.plot(x, y, color='black')

ax.axis([-7, 7, -7, 7])
ax.set_xticks(range(-7, 8))
ax.set_yticks(range(-7, 8))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal', adjustable='box')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/보노보노와 친구들.jpg")

height, width = img.shape[:2]

# 이동 변환 행렬 M
M = np.array([[1, 0, 100], [0, 1, 200], [0, 0, 1]], dtype=float)

# 결과 이미지 초기화
dst = np.copy(img)

for y in range(height):
    for x in range(width):
        # 현재 픽셀 p의 좌표에 이동 변환 행렬 M을 곱하여 이동시킨 좌표를 구한다
        p = np.array([x, y, 1])
        p_ = np.dot(M, p)

        # 이동된 좌표값
        x_, y_ = p_[:2]
        x_ = int(x_)
        y_ = int(y_)

        # 이동된 좌표가 이미지 범위 내에 있는지 확인 후 픽셀 복사
        if 0 <= x_ < width and 0 <= y_ < height:
            dst[y_, x_] = img[y, x]

# 입력 이미지와 결과 이미지를 수평으로 연결하여 하나의 이미지로 만든다
result = cv2.hconcat([img, dst])
cv2_imshow(result)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/보노보노와 친구들.jpg")

height, width = img.shape[:2]

angle = 45
radian = angle * np.pi / 180
c = np.cos(radian)
s = np.sin(radian)

# 회전 변환 행렬 구성
rotation_matrix = np.array([[c, -s, 0], [s, c, 0], [0, 0, 1]], dtype=float)

# 결과 이미지 초기화
dst = np.copy(img)

for y in range(height):
    for x in range(width):
        # 현재 픽셀의 좌표에 회전 변환 행렬을 곱하여 변환된 좌표를 구한다
        new_p = np.array([x, y, 1])
        old_p = np.dot(rotation_matrix, new_p)

        # 변환된 좌표값
        x_, y_ = old_p[:2]
        x_ = int(x_)
        y_ = int(y_)

        # 변환된 좌표가 이미지 범위 내에 있는지 확인 후 픽셀 복사
        if 0 <= x_ < width and 0 <= y_ < height:
            dst[y, x] = img[y_, x_]

# 입력 이미지와 결과 이미지를 수평으로 연결하여 하나의 이미지로 만든다
result = cv2.hconcat([img, dst])
cv2_imshow(result)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/보노보노와 친구들.jpg")

height, width = img.shape[:2]

angle = 45
radian = angle * np.pi / 180
c = np.cos(radian)
s = np.sin(radian)
center_x = width / 2
center_y = height / 2

# 회전 변환 행렬 구성
rotation_matrix = np.array([[c, s, (1 - c) * center_x - s * center_y],
                            [-s, c, s * center_x + (1 - c) * center_y],
                            [0, 0, 1]])

# 결과 이미지 초기화
dst = np.zeros(img.shape, dtype=np.uint8)

for y in range(height):
    for x in range(width):
        # 현재 픽셀의 좌표에 회전 변환 행렬을 곱하여 변환된 좌표를 구한다
        new_p = np.array([x, y, 1])
        old_p = np.dot(rotation_matrix, new_p)

        # 변환된 좌표값
        x_, y_ = old_p[:2]
        x_ = int(x_)
        y_ = int(y_)

        # 변환된 좌표가 이미지 범위 내에 있는지 확인 후 픽셀 복사
        if 0 <= x_ < width and 0 <= y_ < height:
            dst[int(y_), int(x_)] = img[y, x]

# 입력 이미지와 결과 이미지를 수평으로 연결하여 하나의 이미지로 만든다
result = cv2.hconcat([img, dst])
cv2_imshow(result)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/보노보노와 친구들.jpg")

height, width = img.shape[:2]

angle = 45
center = (width // 2, height // 2)  # 중심 좌표는 정수형으로 지정

# 회전 변환 행렬 구하기
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1)

# 이미지 회전
dst = cv2.warpAffine(img, rotation_matrix, (width, height))

# 결과 이미지와 원본 이미지를 수평으로 결합하여 하나의 이미지로 만든다
result = cv2.hconcat([img, dst])

# 결과 이미지 표시
cv2_imshow(result)
cv2.waitKey(0)
```
