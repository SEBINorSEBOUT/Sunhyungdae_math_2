```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a=np.array([5,3,0])
b=np.array([1,2,0])
#외적계산
c=np.cross(a,b)
fig=plt.figure()
ax=fig.add_subplot(1,1,1,projection='3d')
#벡터q를 그린다
ax.quiver(0,0,0,a[0],a[1],a[2],color='black',arrow_length_ratio=0.1)
ax.text(a[0],a[1],a[2],'a',size=15)
#벡터b를 그린다
ax.quiver(0,0,0,b[0],b[1],b[2],color='red',arrow_length_ratio=0.1)
ax.text(b[0],b[1],b[2],'b',size=15)
#벡터a와b의 외적을 그린다
ax.quiver(0,0,0,c[0],c[1],c[2],color='blue',arrow_length_ratio=0.1)
ax.text(c[0],c[1],c[2],'c',size=15)
ax.set_xlim(0,7)
ax.set_ylim(0,7)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_xlabel('Y')
ax.set_xlabel('Z')
ax.view_init(elev=20,azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto',adjustable='box')
plt.show()
```

```python
#항등행렬
import numpy as np
I=np.eye(3)
A=np.arange(9).reshape(3,3)
print(A)
#AI=IA=A임을 확인
print((np.dot(A,I)==A).all())
print((np.dot(I,A)==A).all())
```

```python
#역행렬
import numpy as np
A=np.array([[1,2],[3,4]])
invA=np.linalg.inv(A)
print(invA)
#(A역행렬)A=A(A역행렬)=I임을 확인
#파이썬에서 결과값이 근사값으로 출력된다
#np.round함수를 사용
print(np.dot(A,invA))
print(np.round(np.dot(A,invA)))
print(np.round(np.dot(invA,A)))                           #모르겠담
```
